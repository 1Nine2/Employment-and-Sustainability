```python
sns.set(style="whitegrid")

# Plotting the trends in the data over time
plt.figure(figsize=(12, 6))
plt.plot(df_cleaned['FECHA'], df_cleaned['GRAVA TON'], label='Grava TON', marker='o')
plt.plot(df_cleaned['FECHA'], df_cleaned['GRAVILLA TON'], label='Gravilla TON', marker='o')
plt.plot(df_cleaned['FECHA'], df_cleaned['FINO TON'], label='Fino TON', marker='o')
plt.title('Trends in Material Tons over Time')
plt.xlabel('Date')
plt.ylabel('Tons')
plt.legend()
plt.show()

# Creating a pair plot for pairwise relationships
sns.pairplot(df_cleaned[['GRAVA TON', 'GRAVILLA TON', 'FINO TON']])
plt.suptitle('Pairwise Relationships between Material Tons')
plt.show()

# Creating a bar plot for the total tons of each material
totals = df_cleaned[['GRAVA TON', 'GRAVILLA TON', 'FINO TON']].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=totals.index, y=totals.values)
plt.title('Total Tons of Each Material')
plt.xlabel('Material')
plt.ylabel('Total Tons')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming 'temp' is your DataFrame
# Replace NaN values in 'La Paz' column with the mean
temp['LA PAZ'].fillna(temp['LA PAZ'].mean(), inplace=True)

# Plotting
plt.figure(figsize=(15, 8))

for year in temp['AO'].unique():
    plt.plot(temp[temp['AO'] == year]['MES'], temp[temp['AO'] == year]['LA PAZ'], marker='o', linestyle='-', label=str(year))

plt.title('Temperature Trends in La Paz (1990-2022)')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming 'temp' is your DataFrame
# Replace NaN values in 'La Paz' column with the mean
temp['LA PAZ'].fillna(temp['LA PAZ'].mean(), inplace=True)

# Calculate average temperature for each year
average_temp_by_year = temp.groupby('AO')['LA PAZ'].mean()

# Create a color palette with a different color for each bar
colors = plt.cm.viridis(np.linspace(0, 1, len(average_temp_by_year)))

# Plotting
plt.figure(figsize=(15, 8))
average_temp_by_year.plot(kind='bar', color=colors)
plt.title('Average Temperature in La Paz by Year (1990-2022)')
plt.xlabel('Year')
plt.ylabel('Average Temperature (C)')
plt.grid(axis='y')
plt.show()

```
